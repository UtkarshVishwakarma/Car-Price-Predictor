```python
df = df[df['Price'] < 6000000]
```

```python
df.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler, PowerTransformer, OneHotEncoder
```

```python
trf1 = ColumnTransformer(transformers=[
    ('trf1', StandardScaler(), [2,3])
], remainder='passthrough')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns=['Price'])
y = df['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.2, random_state=178)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
a = trf1.fit_transform(X_train)
```

```python
trf2 = ColumnTransformer(transformers=[
    ('trf1', PowerTransformer(), [0,1])
], remainder='passthrough')
```

```python
a[0]
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name', 'company', 'fuel_type']])
```

```python
cat = ohe.categories_
```

```python
a = trf2.fit_transform(a)
```

```python
a.reshape(-1,1)[0]
```

```python
trf3 = ColumnTransformer(transformers=[
    ('trf1', OneHotEncoder(sparse=True, drop='first', categories=cat), [2,3,4])
], remainder='passthrough')
```

```python
a[0]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
b = a.copy()
```

```python
pipe = Pipeline(steps=[
    ('step1', trf1),
    ('step2', trf2),
    ('step3', trf3),
    ('step4', LinearRegression())
    
])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import Ridge, RANSACRegressor, GammaRegressor, HuberRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor, ExtraTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, BaggingRegressor, GradientBoostingRegressor
```

```python
r2_score(y_test, y_pred)*100
```
